# Task (a)
Use the read.csv() function to read the data into R. Call the
loaded data college. Make sure that you have the directory set
to the correct location for the data.

```r
college = read.csv('./Data/College.csv')
head(college)
```

# Task (b)
rownames, fix, index slicing

```r
rownames(college) = college[,1]
college = college[,-1]
head(college)
```

```r
fix(college) # This will open a new window. Please close it so that the script can continue.
```

# Task (c)
summary, pairs, boxplot, hist

```r
summary(college)
```

```r
pairs(college[, 1:10])
```

```r
boxplot(Outstate ~ Private, college)
```

```r
Elite = rep("No", nrow(college))
Elite[college$Top10perc > 50]="Yes"
Elite = as.factor(Elite)
college = data.frame(college, Elite)
head(college)
```

```r
summary(college['Elite'])
```

```r
par(mfrow=c(2,2))
hist(college$'Apps')
hist(college$'Accept')
hist(college$'Enroll')
hist(college$'Room.Board')
```

```r
for (col in colnames(college)) {
    col_dat = as.vector(college[[col]])
    if (class(col_dat) == 'numeric') {
        hist(col_dat, main=col, xlab=col, col='gray')
    }
}
```
